```python
import pandas as pd
import numpy as np
from pathlib import Path

INPUT_PATH = Path("../data/processed/schools_with_exposure.csv")
OUT_DIR = Path("../data/processed")
OUT_DIR.mkdir(parents=True, exist_ok=True)
```

```python
df = pd.read_csv(INPUT_PATH)

print("Shape:", df.shape)
print("Columns:")
print(df.columns.tolist())
df.head()
```

```python
analysis_cols = [
    "District",
    "School Name",
    "Address",
    "SchoolLevel",
    "Latitude",
    "Longitude",
    "Type",
    "nearest_station_id",
    "nearest_station_name",
    "distance_km",
    "exposure_confidence",
    "PM2.5 (µg/m³)_mean",
    "PM10 (µg/m³)_mean",
    "NO2 (µg/m³)_mean",
    "station_name",
    "network"
]

analysis_df = df[analysis_cols].copy()

print("Analysis table shape:", analysis_df.shape)
analysis_df.head()
```

```python
print("Total schools:", len(analysis_df))
print("Unique districts:", analysis_df["District"].nunique())
print("Unique school types:", analysis_df["Type"].nunique())
print("Unique nearest stations:", analysis_df["nearest_station_id"].nunique())

print("\nMissing values:")
print(analysis_df.isna().sum().sort_values(ascending=False).head(15))
```

##### District Summary Table
This groups all schools by district and calculates average exposure and distance.

```python
district_summary = (
    analysis_df
    .groupby("District", dropna=False)
    .agg(
        school_count=("School Name", "count"),
        avg_pm25=("PM2.5 (µg/m³)_mean", "mean"),
        median_pm25=("PM2.5 (µg/m³)_mean", "median"),
        avg_pm10=("PM10 (µg/m³)_mean", "mean"),
        avg_no2=("NO2 (µg/m³)_mean", "mean"),
        avg_distance_km=("distance_km", "mean"),
        max_distance_km=("distance_km", "max"),
        unique_stations_used=("nearest_station_id", "nunique")
    )
    .reset_index()
)

district_summary["avg_pm25"] = district_summary["avg_pm25"].round(2)
district_summary["median_pm25"] = district_summary["median_pm25"].round(2)
district_summary["avg_pm10"] = district_summary["avg_pm10"].round(2)
district_summary["avg_no2"] = district_summary["avg_no2"].round(2)
district_summary["avg_distance_km"] = district_summary["avg_distance_km"].round(2)
district_summary["max_distance_km"] = district_summary["max_distance_km"].round(2)

district_summary = district_summary.sort_values(
    by="avg_pm25", ascending=False
).reset_index(drop=True)

print("District summary shape:", district_summary.shape)
district_summary.head(10)
```

##### School Type Summary Table
This compares Govt, Unaided, etc.

```python
type_summary = (
    analysis_df
    .groupby("Type", dropna=False)
    .agg(
        school_count=("School Name", "count"),
        avg_pm25=("PM2.5 (µg/m³)_mean", "mean"),
        median_pm25=("PM2.5 (µg/m³)_mean", "median"),
        avg_pm10=("PM10 (µg/m³)_mean", "mean"),
        avg_no2=("NO2 (µg/m³)_mean", "mean"),
        avg_distance_km=("distance_km", "mean"),
        max_distance_km=("distance_km", "max"),
        unique_stations_used=("nearest_station_id", "nunique")
    )
    .reset_index()
)

type_summary["avg_pm25"] = type_summary["avg_pm25"].round(2)
type_summary["median_pm25"] = type_summary["median_pm25"].round(2)
type_summary["avg_pm10"] = type_summary["avg_pm10"].round(2)
type_summary["avg_no2"] = type_summary["avg_no2"].round(2)
type_summary["avg_distance_km"] = type_summary["avg_distance_km"].round(2)
type_summary["max_distance_km"] = type_summary["max_distance_km"].round(2)

type_summary = type_summary.sort_values(
    by="avg_pm25", ascending=False
).reset_index(drop=True)

print("Type summary shape:", type_summary.shape)
type_summary
```

##### School Level Summary Table
This compares Primary / Middle / Secondary / Sr.Secondary.

```python
level_summary = (
    analysis_df
    .groupby("SchoolLevel", dropna=False)
    .agg(
        school_count=("School Name", "count"),
        avg_pm25=("PM2.5 (µg/m³)_mean", "mean"),
        median_pm25=("PM2.5 (µg/m³)_mean", "median"),
        avg_pm10=("PM10 (µg/m³)_mean", "mean"),
        avg_distance_km=("distance_km", "mean"),
        max_distance_km=("distance_km", "max")
    )
    .reset_index()
)

level_summary["avg_pm25"] = level_summary["avg_pm25"].round(2)
level_summary["median_pm25"] = level_summary["median_pm25"].round(2)
level_summary["avg_pm10"] = level_summary["avg_pm10"].round(2)
level_summary["avg_distance_km"] = level_summary["avg_distance_km"].round(2)
level_summary["max_distance_km"] = level_summary["max_distance_km"].round(2)

level_summary = level_summary.sort_values(
    by="avg_pm25", ascending=False
).reset_index(drop=True)

print("Level summary shape:", level_summary.shape)
level_summary
```

##### Top Exposed Schools Ranking
This gives the schools with the highest PM2.5 exposure proxy.

```python
top_exposed_schools = (
    analysis_df
    .sort_values(
        by=["PM2.5 (µg/m³)_mean", "distance_km"],
        ascending=[False, True]
    )
    .reset_index(drop=True)
)

top_exposed_schools = top_exposed_schools[
    [
        "District",
        "School Name",
        "Address",
        "Type",
        "SchoolLevel",
        "nearest_station_name",
        "distance_km",
        "PM2.5 (µg/m³)_mean",
        "PM10 (µg/m³)_mean",
        "NO2 (µg/m³)_mean",
        "exposure_confidence"
    ]
]

top_exposed_schools.head(20)
```

##### Farthest Schools From Monitoring Stations
This helps identify monitoring gaps.

```python
farthest_schools = (
    analysis_df
    .sort_values(by="distance_km", ascending=False)
    .reset_index(drop=True)
)

farthest_schools = farthest_schools[
    [
        "District",
        "School Name",
        "Address",
        "Type",
        "SchoolLevel",
        "nearest_station_name",
        "distance_km",
        "PM2.5 (µg/m³)_mean",
        "exposure_confidence"
    ]
]

farthest_schools.head(20)
```

##### Station Coverage Summary
This tells you which stations are “covering” the largest number of schools.

```python
station_coverage = (
    analysis_df
    .groupby(["nearest_station_id", "nearest_station_name"], dropna=False)
    .agg(
        school_count=("School Name", "count"),
        avg_distance_km=("distance_km", "mean"),
        max_distance_km=("distance_km", "max"),
        avg_pm25=("PM2.5 (µg/m³)_mean", "mean"),
        avg_pm10=("PM10 (µg/m³)_mean", "mean")
    )
    .reset_index()
)

station_coverage["avg_distance_km"] = station_coverage["avg_distance_km"].round(2)
station_coverage["max_distance_km"] = station_coverage["max_distance_km"].round(2)
station_coverage["avg_pm25"] = station_coverage["avg_pm25"].round(2)
station_coverage["avg_pm10"] = station_coverage["avg_pm10"].round(2)

station_coverage = station_coverage.sort_values(
    by="school_count", ascending=False
).reset_index(drop=True)

station_coverage.head(20)
```

##### Confidence Summary
This gives you a quick view of how reliable your nearest-station mapping is.

```python
confidence_summary = (
    analysis_df
    .groupby("exposure_confidence", dropna=False)
    .agg(
        school_count=("School Name", "count"),
        avg_distance_km=("distance_km", "mean"),
        max_distance_km=("distance_km", "max"),
        avg_pm25=("PM2.5 (µg/m³)_mean", "mean")
    )
    .reset_index()
)

confidence_summary["avg_distance_km"] = confidence_summary["avg_distance_km"].round(2)
confidence_summary["max_distance_km"] = confidence_summary["max_distance_km"].round(2)
confidence_summary["avg_pm25"] = confidence_summary["avg_pm25"].round(2)

confidence_summary
```

##### Export All Summary Files

```python
district_summary.to_csv(OUT_DIR / "district_exposure_summary.csv", index=False)
type_summary.to_csv(OUT_DIR / "type_exposure_summary.csv", index=False)
level_summary.to_csv(OUT_DIR / "level_exposure_summary.csv", index=False)
top_exposed_schools.to_csv(OUT_DIR / "top_exposed_schools.csv", index=False)
farthest_schools.to_csv(OUT_DIR / "farthest_schools.csv", index=False)
station_coverage.to_csv(OUT_DIR / "station_coverage_summary.csv", index=False)
confidence_summary.to_csv(OUT_DIR / "exposure_confidence_summary.csv", index=False)

print("Saved files:")
print(OUT_DIR / "district_exposure_summary.csv")
print(OUT_DIR / "type_exposure_summary.csv")
print(OUT_DIR / "level_exposure_summary.csv")
print(OUT_DIR / "top_exposed_schools.csv")
print(OUT_DIR / "farthest_schools.csv")
print(OUT_DIR / "station_coverage_summary.csv")
print(OUT_DIR / "exposure_confidence_summary.csv")
```

##### What Each Output Means
district_exposure_summary.csv = average exposure by district.

type_exposure_summary.csv = average exposure by school type.

level_exposure_summary.csv = average exposure by education level.

top_exposed_schools.csv = schools ranked by highest PM2.5 exposure.

farthest_schools.csv = schools farthest from any station, useful for monitoring-gap analysis.

station_coverage_summary.csv = which stations are assigned the most schools.

exposure_confidence_summary.csv = how many schools are High/Medium/Low confidence based on distance.
