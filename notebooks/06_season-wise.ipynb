```python
def find_first_matching_column(columns, candidates):
    for cand in candidates:
        for col in columns:
            if col.lower() == cand.lower():
                return col
    for cand in candidates:
        for col in columns:
            if cand.lower() in col.lower():
                return col
    return None

station_date_col = find_first_matching_column(
    station_daily.columns,
    ["date", "datetime", "sampling_date", "day", "timestamp"]
)

station_id_col = find_first_matching_column(
    station_daily.columns,
    ["station_id", "station id", "id"]
)

station_name_col = find_first_matching_column(
    station_daily.columns,
    ["station_name", "station name", "station", "location"]
)

pm25_col = find_first_matching_column(
    station_daily.columns,
    ["pm2.5", "pm25"]
)

pm10_col = find_first_matching_column(
    station_daily.columns,
    ["pm10"]
)

no2_col = find_first_matching_column(
    station_daily.columns,
    ["no2"]
)

school_station_id_col = find_first_matching_column(
    schools_exposure.columns,
    ["nearest_station_id", "station_id", "nearest station id"]
)

school_station_name_col = find_first_matching_column(
    schools_exposure.columns,
    ["nearest_station_name", "station_name", "nearest station name", "nearest station"]
)

school_name_col = find_first_matching_column(
    schools_exposure.columns,
    ["school name", "school_name", "name"]
)

district_col = find_first_matching_column(
    schools_exposure.columns,
    ["district"]
)

school_type_col = find_first_matching_column(
    schools_exposure.columns,
    ["type", "school_type", "school type"]
)

print("station_date_col:", station_date_col)
print("station_id_col:", station_id_col)
print("station_name_col:", station_name_col)
print("pm25_col:", pm25_col)
print("pm10_col:", pm10_col)
print("no2_col:", no2_col)
print("school_station_id_col:", school_station_id_col)
print("school_station_name_col:", school_station_name_col)
print("school_name_col:", school_name_col)
print("district_col:", district_col)
print("school_type_col:", school_type_col)
```

```python
station_daily_clean = station_daily.copy()

station_daily_clean["Timestamp"] = pd.to_datetime(
    station_daily_clean["Timestamp"],
    errors="coerce"
)

for col in ["PM2.5 (µg/m³)", "PM10 (µg/m³)", "NO2 (µg/m³)"]:
    station_daily_clean[col] = pd.to_numeric(station_daily_clean[col], errors="coerce")

station_daily_clean = station_daily_clean.dropna(subset=["Timestamp", "station_id"])

print("Cleaned station_daily shape:", station_daily_clean.shape)
station_daily_clean[["Timestamp", "station_id", "station_name", "PM2.5 (µg/m³)", "PM10 (µg/m³)", "NO2 (µg/m³)"]].head()
```

```python
def assign_delhi_season(dt):
    m = dt.month
    if m in [12, 1, 2]:
        return "winter"
    elif m in [3, 4, 5, 6]:
        return "summer"
    elif m in [7, 8, 9]:
        return "monsoon"
    else:
        return "post_monsoon"

station_daily_clean["year"] = station_daily_clean["Timestamp"].dt.year
station_daily_clean["month"] = station_daily_clean["Timestamp"].dt.month
station_daily_clean["season"] = station_daily_clean["Timestamp"].apply(assign_delhi_season)

station_daily_clean[["Timestamp", "year", "month", "season"]].head()
```

```python
print("Years in station data:")
print(station_daily_clean["year"].value_counts().sort_index())

print("\nSeason counts:")
print(station_daily_clean["season"].value_counts())
```

```python
TARGET_YEAR = 2025

station_year = station_daily_clean[station_daily_clean["year"] == TARGET_YEAR].copy()

print("Filtered shape:", station_year.shape)
print(station_year["season"].value_counts())
```

```python
seasonal_station = (
    station_year
    .groupby(["station_id", "station_name", "season"], dropna=False)
    .agg(
        pm25_mean=("PM2.5 (µg/m³)", "mean"),
        pm10_mean=("PM10 (µg/m³)", "mean"),
        no2_mean=("NO2 (µg/m³)", "mean"),
        pm25_median=("PM2.5 (µg/m³)", "median"),
        pm10_median=("PM10 (µg/m³)", "median"),
        no2_median=("NO2 (µg/m³)", "median"),
        record_count=("Timestamp", "count")
    )
    .reset_index()
)

print("seasonal_station shape:", seasonal_station.shape)
seasonal_station.head(10)
```

```python
seasonal_station_wide = seasonal_station.pivot_table(
    index=["station_id", "station_name"],
    columns="season",
    values=[
        "pm25_mean", "pm10_mean", "no2_mean",
        "pm25_median", "pm10_median", "no2_median",
        "record_count"
    ]
)

seasonal_station_wide.columns = [
    f"{metric}_{season}" for metric, season in seasonal_station_wide.columns
]
seasonal_station_wide = seasonal_station_wide.reset_index()

print("seasonal_station_wide shape:", seasonal_station_wide.shape)
seasonal_station_wide.head()
```

```python
seasonal_station_wide = seasonal_station.pivot_table(
    index=["station_id", "station_name"],
    columns="season",
    values=[
        "pm25_mean", "pm10_mean", "no2_mean",
        "pm25_median", "pm10_median", "no2_median",
        "record_count"
    ]
)

seasonal_station_wide.columns = [
    f"{metric}_{season}" for metric, season in seasonal_station_wide.columns
]
seasonal_station_wide = seasonal_station_wide.reset_index()

print("seasonal_station_wide shape:", seasonal_station_wide.shape)
seasonal_station_wide.head()
```

```python
wide = seasonal_station_wide.copy()

if {"pm25_mean_winter", "pm25_mean_monsoon"}.issubset(wide.columns):
    wide["pm25_winter_minus_monsoon"] = wide["pm25_mean_winter"] - wide["pm25_mean_monsoon"]
    wide["pm25_winter_to_monsoon_ratio"] = wide["pm25_mean_winter"] / wide["pm25_mean_monsoon"].replace(0, np.nan)

if {"pm10_mean_winter", "pm10_mean_monsoon"}.issubset(wide.columns):
    wide["pm10_winter_minus_monsoon"] = wide["pm10_mean_winter"] - wide["pm10_mean_monsoon"]
    wide["pm10_winter_to_monsoon_ratio"] = wide["pm10_mean_winter"] / wide["pm10_mean_monsoon"].replace(0, np.nan)

if {"no2_mean_winter", "no2_mean_monsoon"}.issubset(wide.columns):
    wide["no2_winter_minus_monsoon"] = wide["no2_mean_winter"] - wide["no2_mean_monsoon"]
    wide["no2_winter_to_monsoon_ratio"] = wide["no2_mean_winter"] / wide["no2_mean_monsoon"].replace(0, np.nan)

wide.head()
```

```python
schools_seasonal = schools_exposure.merge(
    wide,
    left_on="nearest_station_id",
    right_on="station_id",
    how="left",
    suffixes=("", "_seasonal")
)

print("schools_seasonal shape:", schools_seasonal.shape)
schools_seasonal.head()
```

```python
AVG_ENROLMENT_DELHI = 808

for season in ["winter", "summer", "monsoon", "post_monsoon"]:
    pm25_col_name = f"pm25_mean_{season}"
    if pm25_col_name in schools_seasonal.columns:
        schools_seasonal[f"students_affected_{season}_est"] = AVG_ENROLMENT_DELHI

schools_seasonal.head()
```

```python
district_agg = (
    schools_seasonal
    .groupby("District", dropna=False)
    .agg(
        school_count=("School Name", "count"),
        pm25_mean_winter=("pm25_mean_winter", "mean"),
        pm25_mean_summer=("pm25_mean_summer", "mean"),
        pm25_mean_monsoon=("pm25_mean_monsoon", "mean"),
        pm25_mean_post_monsoon=("pm25_mean_post_monsoon", "mean"),
        pm10_mean_winter=("pm10_mean_winter", "mean"),
        pm10_mean_summer=("pm10_mean_summer", "mean"),
        pm10_mean_monsoon=("pm10_mean_monsoon", "mean"),
        pm10_mean_post_monsoon=("pm10_mean_post_monsoon", "mean"),
        no2_mean_winter=("no2_mean_winter", "mean"),
        no2_mean_summer=("no2_mean_summer", "mean"),
        no2_mean_monsoon=("no2_mean_monsoon", "mean"),
        no2_mean_post_monsoon=("no2_mean_post_monsoon", "mean")
    )
    .reset_index()
)

district_agg["students_affected_est"] = district_agg["school_count"] * AVG_ENROLMENT_DELHI

if {"pm25_mean_winter", "pm25_mean_monsoon"}.issubset(district_agg.columns):
    district_agg["pm25_winter_minus_monsoon"] = district_agg["pm25_mean_winter"] - district_agg["pm25_mean_monsoon"]
    district_agg["pm25_winter_to_monsoon_ratio"] = district_agg["pm25_mean_winter"] / district_agg["pm25_mean_monsoon"].replace(0, np.nan)

district_agg.head()
```

```python
top_winter_pm25_schools = (
    schools_seasonal
    .sort_values("pm25_mean_winter", ascending=False)
    [[
        "District", "School Name", "Type", "nearest_station_name", "distance_km",
        "pm25_mean_winter", "pm25_mean_monsoon",
        "pm25_winter_minus_monsoon", "pm25_winter_to_monsoon_ratio"
    ]]
    .head(50)
)

top_pm25_seasonal_jump_schools = (
    schools_seasonal
    .sort_values("pm25_winter_minus_monsoon", ascending=False)
    [[
        "District", "School Name", "Type", "nearest_station_name", "distance_km",
        "pm25_mean_winter", "pm25_mean_monsoon",
        "pm25_winter_minus_monsoon", "pm25_winter_to_monsoon_ratio"
    ]]
    .head(50)
)

top_winter_pm25_schools.head(), top_pm25_seasonal_jump_schools.head()
```
